# Time Series NDVI via Agromonitoring.com

define a polygon:

```python
body1 = {
   "name":"Polygon Sample",
   "geo_json":{
      "type":"Feature",
      "properties":{

      },
      "geometry":{
         "type":"Polygon",
         "coordinates":[
            [
               [-94.4545917478666, 41.9792090154671], 
                [-94.4545448033213, 41.9757220431519], 
                [-94.4450066084548, 41.9757090969481], 
                [-94.4450437851949, 41.9792826686391], 
                [-94.4545917478666, 41.9792090154671]
            ]
         ]
      }
   }
}
```

register the polygon:

```python
import requests

url = 'http://api.agromonitoring.com/agro/1.0/polygons?appid=4a8757d4f5eda548879a80893a7cefa1'



response = requests.post(url, json=body1)

response.content
```

get historical time series data

```python
id = '64da6680287b0e1b44fce937'
startdate = '1660513750' # UNIX date, 2022-08-14
enddate = '1691963350' # UNIX date, 2023-08-14
```

```python
url = f'http://api.agromonitoring.com/agro/1.0/ndvi/history?start={startdate}&end={enddate}&polyid={id}&appid=4a8757d4f5eda548879a80893a7cefa1'
```

```python
response = requests.get(url)
response.content
```

```python
import pandas as pd

l = response.json()
dates = []
type = []
values = []
for i in l:
    dates.append(i['dt'])
    type.append(i['type'])
    values.append(i['data']['median'])
df = pd.DataFrame({'date': dates, 'type': type, 'value': values})
df['date'] = pd.to_datetime(df['date'], unit='s')
df['value'] = df['value'].astype(float)
df.head(10)
```
